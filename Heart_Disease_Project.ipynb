# Clustering Using K-means

1.Assign the documents randomly to k bins
2.Compute the location of the centroid of each bin.
3.Compute the distance between each document and each centroid
4.Assign each document to the bin corresponding to the centroid closest to it.
5.Stop if no document is moved to a new bin, else go to step 2.

```r
library(gridExtra)
library(factoextra)
```

```r
data3 <- scale(data2,center=apply(data2,2,median),scale = TRUE) #scaling feature
```

Scaling the features makes the flow of gradient descent smooth and helps algorithms quickly reach the minima of the cost function. Without scaling features, the algorithm may be biased toward the feature which has values higher in magnitude.
(K-means) as these are very sensitive to the range of the data points

```r
distance <- get_dist(data3) #Computes a distance matrix between the rows of a data matrix & returns an object of class "dist"
head(distance)
```

```r
fviz_dist(distance, gradient = list(low = "#00AFBB", mid = "white", high = "#FC4E07"))
#Visualizes a distance matrix  & returns a ggplot2 
# gradient is a list containing three elements specifying the colors for low, mid and high values in the ordered dissimilarity 
#image. The element "mid" can take the value of NULL.
```

# K-Medoids

```r
pam(data3, k=2, metric = "euclidean", stand = FALSE)
#pam: partitioning around medians to perform k-medoids clustering
```

```r
fviz_nbclust(data3, pam, method = "wss")
```

gap_stat: to determine the optimal number of clusters is to use a metric known as the gap statistic, which compares the total intra-cluster variation for different values of k with their expected values for a distribution with no clustering.

```r
#calculate gap statistic based on number of clusters
gap_stat <- clusGap(data3,
                    FUN = pam,
                    K.max = 10, #max clusters to consider
                    B = 50) #total bootstrapped iterations

#plot number of clusters vs. gap statistic
fviz_gap_stat(gap_stat)
```

```r
kmed <- pam(data3, k = 2)
```
